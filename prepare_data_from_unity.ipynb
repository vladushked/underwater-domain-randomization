# Prepare data from unity

```python
import cv2
import glob
import os
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path
import json
import xml.etree.ElementTree as tree
import shutil
```

```python
images_path = "/home/vladushked/Documents/data/unity/dataset/images/round_1"
labels_path = "/home/vladushked/Documents/data/unity/dataset/labels/round_1"
images = glob.glob(os.path.join(images_path, "*.png"))
labels = glob.glob(os.path.join(labels_path, "*.txt"))
images.sort()
labels.sort()
label_names = ['gate', 'red_flare', 'yellow_flare', 'red_bowl1', 'red_bowl2', 'red_bowl3', 'blue_bowl', 'mat']
out_label_names = ['gate', 'red_flare', 'yellow_flare', 'red_bowl', 'blue_bowl', 'mat']
```

```python
out_labels = "/home/vladushked/Documents/data/unity/dataset/labels/common"
```

```python
for path in labels:
    filename = os.path.split(path)[-1]
    array = np.genfromtxt(path,dtype='str')
    if array.size == 0:
        shutil.copy2(path, out_labels)
        continue
    if array.ndim == 1:
        array = [array]
    with open(os.path.join(out_labels, filename), "a") as output:
        for line in array:
            label = label_names[int(line[0])]
            if label == 'red_bowl1' or label == 'red_bowl2' or label == 'red_bowl3':
                label = "red_bowl"
            idx = out_label_names.index(label)
            if idx > 5:
                print(filename)
            x_center = float(line[1].replace(',','.')) 
            y_center = float(line[2].replace(',','.'))
            width = float(line[3].replace(',','.'))
            height = float(line[4].replace(',','.'))
            out = [str(idx), "%.6f" % x_center, "%.6f" % y_center, "%.6f" % width, "%.6f" % height]
            out_str = " ".join(out) + "\n"
            output.write(out_str)
```

```python
for i in range(0, len(images), 25):
    image = cv2.imread(images[i])
    heigth, width, _ = image.shape
    txt = np.genfromtxt(labels[i],dtype='str')
    print(txt)
    lines = np.genfromtxt(labels[i],dtype='str')
    if lines.ndim == 1:
        lines = [lines]
    for line in lines:
        label = int(line[0])
        x_center = int(float(line[1].replace(',','.')) * width)
        y_center = int(float(line[2].replace(',','.')) * heigth)
        box_width = int(float(line[3].replace(',','.')) * width)
        box_height = int(float(line[4].replace(',','.')) * heigth)
        cv2.rectangle(image,(x_center - box_width // 2, y_center - box_height // 2),(x_center + box_width // 2, y_center + box_height // 2),(0,255,0),3)
        cv2.circle(image, (x_center,y_center), radius=0, color=(0, 0, 255), thickness=20)
        cv2.putText(image,label_names[label],(x_center,y_center), cv2.FONT_HERSHEY_SIMPLEX, 1,(255,255,255),1,cv2.LINE_AA)
    cv2.imwrite("output/out%d.jpg" % (i), image)
```

```python
for image in images:
    img = cv2.imread(image)
    filename = os.path.split(image)[-1].split(".")[0] + ".jpg"
    print(filename)
    cv2.imwrite(os.path.join("/home/vladushked/Documents/research/Yolo2Pascal-annotation-conversion/common", filename), img)
```

## save refined labels to other rounds and shuffle

```python
import cv2
import glob
import os
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path
import json
import xml.etree.ElementTree as tree
import shutil
```

```python
image_source = "/home/vladushked/Documents/data/unity/dataset/images"
round_1_shuffled = "/home/vladushked/Documents/data/unity/round_1_shuffled"
dataset_dir = "/home/vladushked/Documents/data/unity"
```

```python
train_labels = glob.glob(os.path.join(round_1_shuffled, "train/labels", "*.txt"))
eval_labels = glob.glob(os.path.join(round_1_shuffled, "eval/labels", "*.txt"))
```

```python
out_rounds = [2,3]
for round in out_rounds:
    output_round_dir = "round_" + str(round) + "_shuffled"
    output_round_dir = os.path.join(dataset_dir, output_round_dir)
    Path(output_round_dir).mkdir(parents=True, exist_ok=True)

    train_dir = os.path.join(output_round_dir, "train")
    eval_dir = os.path.join(output_round_dir, "eval")
    Path(train_dir).mkdir(parents=True, exist_ok=True)
    Path(eval_dir).mkdir(parents=True, exist_ok=True)

    # train
    Path(os.path.join(train_dir, "labels")).mkdir(parents=True, exist_ok=True)
    Path(os.path.join(train_dir, "images")).mkdir(parents=True, exist_ok=True)
    for label_path in train_labels:
        label_name = os.path.split(label_path)[-1]
        label_name_splitted = label_name.split("_")
        label_name_splitted[2] = str(round)
        new_label_name = "_".join(label_name_splitted)
        shutil.copy2(label_path, os.path.join(train_dir, "labels", new_label_name))
        image_name = new_label_name.split(".")[0] + ".png"
        shutil.copy2(os.path.join(image_source, "round_" + str(round), image_name), os.path.join(train_dir, "images", image_name))

    # eval
    Path(os.path.join(eval_dir, "labels")).mkdir(parents=True, exist_ok=True)
    Path(os.path.join(eval_dir, "images")).mkdir(parents=True, exist_ok=True)
    for label_path in eval_labels:
        label_name = os.path.split(label_path)[-1]
        label_name_splitted = label_name.split("_")
        label_name_splitted[2] = str(round)
        new_label_name = "_".join(label_name_splitted)
        shutil.copy2(label_path, os.path.join(eval_dir, "labels", new_label_name))
        image_name = new_label_name.split(".")[0] + ".png"
        shutil.copy2(os.path.join(image_source, "round_" + str(round), image_name), os.path.join(eval_dir, "images", image_name))
```
